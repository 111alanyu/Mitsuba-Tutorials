```python
import mitsuba as mi
import matplotlib.pyplot as plt
%matplotlib inline

#this sets variant so all uses 'scalar_rgb' for future mi function calls
mi.set_variant('scalar_rgb') 
```

```python
assets = "/Users/alanyu/Documents/Programming/EECE-194/Mitsuba3/scenes/"

scene = mi.load_file(assets + "cbox.xml")
image = mi.render(scene)

mi.Bitmap(image)
mi.util.write_bitmap("cbox_render.png", image)
```

```python
#Rendering the scene from a different view point

cam2 = mi.load_dict({
    'type': 'perspective',
    'to_world': mi.ScalarTransform4f.look_at(
        target=[-5.0, 0.0, 0.0],
        origin=[1, 0, 0],
        up = [0.0,1.0,2.0]
    ),
    'film' : {
        'type': 'hdrfilm'
    }
})

image2 = mi.render(scene, sensor = cam2)
mi.Bitmap(image2)

f, axarr = plt.subplots(2,1) 
axarr[0].imshow(image)
axarr[1].imshow(image2)

mi.util.write_bitmap("cbox_render_2.png", image2)
```
